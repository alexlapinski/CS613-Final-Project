```python
import os
import sys
import pandas as pd

src_dir = os.path.join(os.getcwd(), os.pardir)
sys.path.append(src_dir)

#from data_quality import data.hdd

raw_data_dir = os.path.join(src_dir, 'data', 'raw')

raw_df = pd.read_csv(os.path.join(raw_data_dir, 'hdd', '2016-04-01.csv'), index_col=[0, 1, 2], dtype=str)

# Drop the unneeded column from the index
new_index = raw_df.index.droplevel()
raw_df.index = new_index

# Move the 'failure' column to 'label' (Possible values of 'failure' column are '0' and '1')
# '1' indicates this was the last day before the drive failed.
raw_df['label'] = -1
raw_df[raw_df['failure'] == 0] = 1
raw_df.drop('failure', axis=1, inplace=True)

# Convert our 'raw' columns from hexadecimal to decimal
raw_column_names = [col for col in raw_df.columns if col.endswith('_raw')]


def convert_base(x):
        try:
                return int(x, 16)
        except:
                pass
        return x

for col in raw_column_names:
        raw_df[col] = raw_df[col].apply(convert_base)

raw_df
```
