```python
import os
import sys
import pandas as pd

src_dir = os.path.join(os.getcwd(), os.pardir)
sys.path.append(src_dir)

#from data_quality import data.hdd

raw_data_dir = os.path.join(src_dir, 'data', 'raw')

raw_df = pd.read_csv(os.path.join(raw_data_dir, 'hdd', '2016-04-01.csv'), index_col=[0, 1, 2])

# Drop the unneeded column from the index
new_index = raw_df.index.droplevel()
raw_df.index = new_index

# For each column, let's see what the percentage of NaN values are, 
# we may want to discard the column if the percentage is above a threshold

percent_nan = (raw_df.isnull().sum() / raw_df.size)

print "Columns with Nan Values > 1%"
print "Num Columns: {0}".format(percent_nan[percent_nan > 0.01].size)
print "Total Number of Columns: {0}".format(raw_df.columns.size)
print "Column Name             Percent Missing"
print percent_nan[percent_nan > 0.01]


# Based on this, we don't need to cut our columns
```
